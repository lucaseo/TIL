### 모수지정

확률 분포 객체 생성 시, 분포의 형상을 구체적으로 지정해주는 모수(parameter) 이정.

`loc` = 분포의 기대값
`scale` = 분포의 표준편차

```python
tmp = sp.stats.norm(loc=1 , scale=2)
```

### 확률 분포 메서드

- `pdf` : 확률 밀도 함수
- `pmf` : 확률 질량 함수
- `cdf` : 누적 분포 함수
- `rvs` : 랜덤 샘플 생성

```python
# pdf

xx = np.linspace(-10, 10, 100)
pdf = tmp.pdf(xx)
plt.plot(xx, pdf)
plt.show()
```

```python
# cdf

xx = np.linspace(-10, 10, 100)
cdf = tmp.cdf(xx)
plt.plot(xx, cdf)
plt.show()
```

### random sample 생성
- `size` : 생성된 랜덤 샘플의 크기
- `random_state` : 랜덤 샘플 생성 시, 시드값

```python
tmp.rvs(size=(4, 6), random_state=0)
```
