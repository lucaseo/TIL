# 스튜던트 t 분포

### Fat tail?
- 정규분포보다 양 끝단의 비중이 더 큰 데이터.
- 이러한 데이터 모형에 적합한 것이 스튜던트 t분포

(PDF 수식) 자유도 $\nu$를 추가적인 파라미터로 가진다.

$$f(x) = \frac{\Gamma\left(\frac{\nu+1}{2}\right)} {\sqrt{\nu\pi}\Gamma\left(\frac{\nu}{2}\right)} \left(1+\frac{(x-\mu)^2}{\nu\sigma^2} \right)^{-\frac{\nu+1}{2}}$$

# t 분포 그리기

```python
import numpy as np
import scipy as sp
from scipy import stats
import matplotlib as mpt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
xx = np.linspace(-4, 4, 100)
for df in [1, 2, 5, 10, 20]:
    rv = sp.stats.t(df=df)
    plt.plot(xx, rv.pdf(xx), label=("student-t (dof=%d)" % df))
plt.plot(xx, sp.stats.norm().pdf(xx), label="Normal", lw=5, alpha=0.5)
plt.legend()
plt.show()
```

- 자유도가 작으면 $\mathcal{N}$보다 분산이 크고 fat tail을 보인다. 
- 자유도가 증가할 수록 정규분포로 수렴

### 모멘트

기대값 : 

$$\text{E}[X] = \mu$$

분산 : ($\nu$ > 2일 때만 적용 $\nu$ = 1, 2일 때는 분산이 무한대)

$$\text{Var}[X] = \dfrac{\nu \sigma^2}{\nu-2}$$


# 정규분포의 샘플 평균이 이루는 분포 (중요)

샘플의 수가 무한대가 아닌 유한한 경우?

기존 $\sigma^2$를 상수로 정규화한 샘플 평균은 정규분포 

$$\frac{\bar{x} - \mu}{\dfrac{\sigma}{\sqrt{n}}} \sim \mathcal{N}(x; \mu, \sigma^2)$$

샘플 평균을 샘플 분산 $s$로 정규화한 값은 t분포

$$\frac{\bar{x} - \mu}{\dfrac{s}{\sqrt{n}}} \sim t(x; \mu, \sigma^2, n - 1)$$

### 통계량(statistics)이란 ? 

복수의 샘플데이터집합을 연산해서 구한 숫자가 통계량
(예)
- 정규화된 샘플 표본
- 평균, 분산 등등 
- t분포는 정규화된 샘플표본이라는 통계량이 따르는 분포
