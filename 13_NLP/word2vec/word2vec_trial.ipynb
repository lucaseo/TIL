```python
import re
import pandas as pd
from konlpy.tag import Twitter 

from tqdm import tqdm_notebook
```

```python
hangul = re.compile('[^ ㄱ-ㅣ가-힣]+')
```

```python
tagger = Twitter()
```

```python
data = pd.read_csv('../../../../00_dataset_archive/nlp/news_data_cleaned.csv')
```

```python
data[['content']][:10000].head()
```

```python
data_tmp = data[['content']][:10000]
```

```python
docs = []

for article in tqdm_notebook(data_tmp['content']):
    article = hangul.sub(' ', article)
    article = tagger.morphs(article, stem=True)
    article = [x for x in article if len(x) > 1]
    docs.append(article)
```

```python
docs[0]
```

```python
from gensim.models import word2vec
```

```python
import logging
logging.basicConfig(
    format='%(asctime)s : %(levelname)s : %(message)s', 
    level=logging.INFO)
```

```python
num_features = 500
min_word_count = 10
num_workers = 1
context = 15
downsampling = 1e-3
```

```python
model = word2vec.Word2Vec(sentences=docs, 
                         workers=num_workers,
                         size=num_features,
                         min_count=min_word_count,
                         window=context,
                         sample=downsampling
                         )
```

```python
model
```

```python
model.init_sims(replace=True)
```

```python
model_name = "word2vec trial"
model.save(model_name)
```
