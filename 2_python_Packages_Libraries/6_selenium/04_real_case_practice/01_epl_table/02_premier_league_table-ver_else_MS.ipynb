```python
from selenium import webdriver
import pandas as pd
```

```python
df = pd.DataFrame(columns=['Ranking', 'Club', 'Pl', 'W', 'D', 'L', 'GD', 'Pts'])
```

```python
def epl_table():
    
    driver = webdriver.Chrome()
    driver.get('https://www.premierleague.com/tables')
    
    try:
        driver.find_element_by_css_selector("#advertClose").click()
    except:
        print("no layer")
    
    teams = driver.find_elements_by_css_selector(".allTablesContainer table > tbody > tr:not(.expandable)")
    
    
    
    for idx, team in enumerate(teams):
        
        ranking = team.find_element_by_css_selector("td > .value")
        club = team.find_element_by_css_selector(".team .long")
        play = team.find_element_by_css_selector("td:nth-child(4)")
        win = team.find_element_by_css_selector("td:nth-child(5)")
        draw = team.find_element_by_css_selector("td:nth-child(6)")
        lose = team.find_element_by_css_selector("td:nth-child(7)")
        goal = team.find_element_by_css_selector("td:nth-child(10)")
        pts = team.find_element_by_css_selector(".points")
        data = {"Ranking" : ranking.text, 
                "Club" : club.text, 
                "Pl" : play.text,
                "W" : win.text,
                "D" : draw.text,
                "L" : lose.text,
                "GD": goal.text, 
                "Pts": pts.text }

        df.loc[len(df)] = data
        
        if idx == 19:
            break
    
    df.set_index('Ranking', inplace=True)
    df.to_csv('epl_table')
    
    return "File generated !"
```

```python
epl_table()
```

```python
data = pd.read_csv('epl_table', index_col = 0)
```
